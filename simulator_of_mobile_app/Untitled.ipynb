```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv', names=['error_code', 'error_message', 'severity', 'log_location', 'mode', 'model', 'graphics', 'session_id', 'sdkv', 'test_mode', 'flow_id', 'flow_type', 'sdk_date', 'publisher_id', 'game_id', 'bundle_id', 'appv', 'language', 'os', 'adv_id', 'gdpr', 'ccpa', 'country_code', 'date'])[1:]
data['date'] = pd.to_datetime(data['date'], unit='s')
data.sort_values(by="date")
```

```python
data[(data['severity'] == 'Error') & (data['error_code'] == 203)]
```

```python
data['minute'] = data['date'].apply(lambda ts: ts.minute)
(data[(data['severity'] == 'Error') & (data['error_code'] == 203)].groupby('minute').count()['error_code'] > 10).any()
```

```python
(data.groupby(['bundle_id', 'hour']).count()['error_code'] > 10).groupby('bundle_id').any().index.to_list()
```

```python
data['hour'] = data['date'].apply(lambda ts: ts.hour)
((data.groupby(['bundle_id', 'hour']).count()['error_code'] > 10).groupby('bundle_id').any()).any()
```
